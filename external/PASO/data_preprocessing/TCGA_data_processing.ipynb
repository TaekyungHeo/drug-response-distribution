```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('tcga_result/TCGA_drug_sensitivity_MixedSet_train.csv')
df_test = pd.read_csv('tcga_result/TCGA_drug_sensitivity_MixedSet_test.csv')
print(df_train.shape)
print(df_test.shape)
# Merge df_train and df_test
df_drug_sensitivity = pd.concat([df_train, df_test])
print(df_drug_sensitivity.shape)
```

```python
# Count the number of different values in the bcr_patient_barcode column; the number of different values in the drug column; the number of different values in the project column in df_drug_sensitivity.

print('patients',df_drug_sensitivity['bcr_patient_barcode'].nunique())
print('drugs',df_drug_sensitivity['drug'].nunique())
print('projects',df_drug_sensitivity['project'].nunique())
```

```python
df_drug_smiles = pd.read_csv('tcga_result/TCGA_Drug_Name_SMILES.csv')[['CID','DRUG_NAME', 'Canonical_SMILES']]
print(df_drug_smiles.shape)
df_drug_smiles.head()
```

```python
# df_drug_sensitivity merged with df_drug_smiles based on the former drug_name column and the latter DRUG_NAME column

df_drug_sensitivity = pd.merge(df_drug_sensitivity, df_drug_smiles, left_on='drug_name', right_on='DRUG_NAME', how='left')
print(df_drug_sensitivity.shape)
df_drug_sensitivity.head()
```

## read TCGA drug sensitivity data

```python
df_drug_sensitivity = pd.read_csv('tcga_result/TCGA_drug_sensitivity.csv')
print(df_drug_sensitivity.shape)
df_drug_sensitivity.head()
```

```python
# df_drug_sensitivity Group the DRUG_NAMEs according to the CID column, count the number of different DRUG_NAMEs in each group, and change all the DRUG_NAMEs to the one with the most occurrences.

df_drug_sensitivity['DRUG_NAME'] = df_drug_sensitivity.groupby('CID')['DRUG_NAME'].transform(lambda x: x.value_counts().index[0])
# df_drug_sensitivity = df_drug_sensitivity.drop_duplicates()
print(df_drug_sensitivity.shape)
df_drug_sensitivity.head()
```

```python
# Count the number of different values in the bcr_patient_barcode column; count the number of different values in the drug_name column.
print('patients',df_drug_sensitivity['bcr_patient_barcode'].nunique())
print('drugs',df_drug_sensitivity['DRUG_NAME'].nunique())
```

```python
# tocsv
# df_drug_sensitivity.to_csv('tcga_result/TCGA_drug_sensitivity_filtered.csv', index=False)
```

## TCGA drug sensitivity data -> SMILES DATA 

```python
# Extract the DRUG_NAME and Canonical_SMILES columns from df_drug_sensitivity, de-duplicated
df_drug_smiles = df_drug_sensitivity[['CID','DRUG_NAME', 'Canonical_SMILES']].drop_duplicates()
# Sorted by DRUG_NAME columns
df_drug_smiles = df_drug_smiles.sort_values(by='DRUG_NAME')
print(df_drug_smiles.shape)
df_drug_smiles.head()
```

## drug sensitivity data -> addition of CancerType/project

```python
df_clinical = pd.read_csv('tcga_result/TCGA_clinical.csv')[['bcr_patient_barcode','project']]
df_clinical = df_clinical.drop_duplicates()
print(df_clinical.shape)
df_clinical.head()
```

```python
df_drug_sensitivity = pd.read_csv('tcga_result/TCGA_drug_sensitivity.csv')
print(df_drug_sensitivity.shape)
df_drug_sensitivity.head()
```

```python
# df_drug_sensitivity is merged with df_clinical, based on the former bcr_patient_barcode column and the latter bcr_patient_barcode column.
df_drug_sensitivity = pd.merge(df_drug_sensitivity, df_clinical, on='bcr_patient_barcode', how='left')
print(df_drug_sensitivity.shape)
df_drug_sensitivity.head()
```

```python
# tocsv
# df_drug_sensitivity.to_csv('tcga_result/TCGA_drug_sensitivity_with_project.csv', index=False)
```

```python
df_drug_sensitivity_all = pd.read_csv('tcga_result/TCGA_drug_sensitivity.csv')[['drug','bcr_patient_barcode','response','measure_of_response']]
print(df_drug_sensitivity_all.shape)
df_drug_sensitivity_all.head()
```

```python
df_drug_sensitivity_train = pd.read_csv('tcga_result/TCGA_drug_sensitivity_MixedSet_train.csv',index_col=0)
df_drug_sensitivity_test = pd.read_csv('tcga_result/TCGA_drug_sensitivity_MixedSet_test.csv',index_col=0)
print(df_drug_sensitivity_train.shape)
print(df_drug_sensitivity_test.shape)
```

```python
df_predictions = pd.read_csv('tcga_result/TCGA_drug_sensitivity_MixedSet_predictions.csv')
print(df_predictions.shape)
```

```python
# Merge df_drug_sensitivity_train, df_drug_sensitivity_test and df_drug_sensitivity_all based on drug and bcr_patient_barcode columns
df_predictions = pd.merge(df_predictions, df_drug_sensitivity_all, on=['drug','bcr_patient_barcode','response'], how='left')
print(df_predictions.shape)
df_predictions = df_predictions.drop_duplicates()
print(df_predictions.shape)
df_predictions.head()
```

```python
# tocsv
df_predictions.to_csv('tcga_result/TCGA_drug_sensitivity_MixedSet_predictions_with_response.csv', index=False)
```

```python
df_gep_bcr_patient_barcode = pd.read_csv('tcga_result/TCGA_GEP_Wilcoxon_Test_Analysis_Log10_P_value_C2_KEGG_MEDICUS.csv')[['bcr_patient_barcode']]
print(df_gep_bcr_patient_barcode.shape)
df_gep_bcr_patient_barcode.head()
```

```python
# Filter out rows from df_drug_sensitivity_train where the bcr_patient_barcode column does not exist in the bcr_patient_barcode column of df_gep_bcr_patient_barcode.
df_drug_sensitivity_train = df_drug_sensitivity_train[~df_drug_sensitivity_train['bcr_patient_barcode'].isin(df_gep_bcr_patient_barcode['bcr_patient_barcode'])]
print(df_drug_sensitivity_train.shape)
df_drug_sensitivity_train.head()
```

```python
# Filter out the rows in df_drug_sensitivity_test where the bcr_patient_barcode column does not exist in the bcr_patient_barcode column of df_gep.
df_drug_sensitivity_test = df_drug_sensitivity_test[~df_drug_sensitivity_test['bcr_patient_barcode'].isin(df_gep_bcr_patient_barcode['bcr_patient_barcode'])]
print(df_drug_sensitivity_test.shape)
df_drug_sensitivity_test.head()
```

```python
# Merge df_drug_sensitivity_train and df_drug_sensitivity_test.
df_drug_sensitivity = pd.concat([df_drug_sensitivity_train, df_drug_sensitivity_test])[['bcr_patient_barcode']]
print(df_drug_sensitivity.shape)
df_drug_sensitivity.head()
```

```python
# Remove duplicates.
df_drug_sensitivity = df_drug_sensitivity.drop_duplicates()
print(df_drug_sensitivity.shape)
df_drug_sensitivity.head()
```

```python
# 去除df_drug_sensitivity_train和df_drug_sensitivity_test中bcr_patient_barcode列与df_drug_sensitivity中bcr_patient_barcode存在的行
df_drug_sensitivity_train = df_drug_sensitivity_train[~df_drug_sensitivity_train['bcr_patient_barcode'].isin(df_drug_sensitivity['bcr_patient_barcode'])]
df_drug_sensitivity_test = df_drug_sensitivity_test[~df_drug_sensitivity_test['bcr_patient_barcode'].isin(df_drug_sensitivity['bcr_patient_barcode'])]
print(df_drug_sensitivity_train.shape)
print(df_drug_sensitivity_test.shape)
```

```python
# tocsv
df_drug_sensitivity_train.to_csv('tcga_result/TCGA_drug_sensitivity_MixedSet_train.csv', index=False)
df_drug_sensitivity_test.to_csv('tcga_result/TCGA_drug_sensitivity_MixedSet_test.csv', index=False)
```

```python
# Merge df_drug_sensitivity_train and df_drug_sensitivity_test.
df_drug_sensitivity = pd.concat([df_drug_sensitivity_train, df_drug_sensitivity_test])
print(df_drug_sensitivity.shape)
df_drug_sensitivity.head()
```

```python
# Count the number of patients, drugs, and projects.
print('patients',df_drug_sensitivity['bcr_patient_barcode'].nunique())
print('drugs',df_drug_sensitivity['drug'].nunique())
print('projects',df_drug_sensitivity['project'].nunique())
```

```python
# Sort by drug and bcr_patient_barcode, then reset the index.
df_drug_sensitivity = df_drug_sensitivity.sort_values(by=['bcr_patient_barcode'])
df_drug_sensitivity = df_drug_sensitivity.reset_index(drop=True)
df_drug_sensitivity.to_csv('tcga_result/TCGA_drug_sensitivity_filtered.csv', index=False)
```
